```python
import requests
import pathlib
```

```python
fairscape_root = "https://fairscape.net/"
```

```python
uploadedCrates = '/Users/mal8ch-admin/Box Sync/cm4ai-data/Dataverse-Uploaded/'
#rocrate_path = '/Users/mal8ch-admin/Box Sync/cm4ai-data/Dataverse-Uploaded/4.cm4ai_chromatin_mda-mb-468_untreated_hierarchy_initialrun0.1alpha.zip'
#testCrate = {'crate': ('4.cm4ai_chromatin_mda-mb-468_untreated_hierarchy_initialrun0.1alpha.zip', open(rocrate_path, 'rb'), 'application/zip')}

allUploadedCrates = list(pathlib.Path(uploadedCrates).glob('*.zip'))
```

```python
def uploadCrate(roCratePath: pathlib.Path):

    fileName = roCratePath.name
    crateUpload = {'crate': (fileName, open(str(roCratePath), 'rb'), 'application/zip')}
    
    uploadAsyncResponse = requests.post(
        fairscape_root + 'rocrate/upload-async',
        files=crateUpload
    )
    return uploadAsyncResponse
```

```python
def checkUploadStatus(uploadResponseJSON):
    uploadRequestUUID = uploadResponseJSON['transactionFolder']

    uploadStatusResponse = requests.get(
        fairscape_root + 'rocrate/upload-async/status/' + uploadRequestUUID
    )
    return uploadStatusResponse
```
